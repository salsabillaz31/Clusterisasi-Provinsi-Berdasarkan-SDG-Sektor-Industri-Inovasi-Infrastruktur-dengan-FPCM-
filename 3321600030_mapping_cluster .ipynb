```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(databaru)
df_scaled = scaler.transform(databaru)
databaru = pd.DataFrame(df_scaled, columns=['F1', 'F2', 'F3', 'F4','F5', 'F6',
                                             'F7', 'F8', 'F9'])
databaru
```

```python
import numpy as np
from scipy.spatial.distance import cdist

class FPCM:
    def __init__(self, n_clusters, m=2, epsilon=0.01, max_iter=100):
        self.n_clusters = n_clusters  # Number of clusters
        self.m = m  # Fuzziness parameter
        self.epsilon = epsilon  # Stopping criterion
        self.max_iter = max_iter  # Maximum number of iterations

    def fit(self, X):
        self.X = X  # Input data matrix
        self.n_samples, self.n_features = X.shape

        # Randomly initialize the cluster centers
        self.centers = np.random.rand(self.n_clusters, self.n_features)

        # Initialize the membership matrix
        self.membership_mat = np.random.rand(self.n_samples, self.n_clusters)
        self.membership_mat /= np.sum(self.membership_mat, axis=1)[:, np.newaxis]

        # Main loop
        for _ in range(self.max_iter):
            # Calculate the cluster centers
            self.calculate_centers()

            # Update the membership matrix
            self.update_membership_mat()

            # Check for convergence
            if np.linalg.norm(self.membership_mat - self.old_membership_mat) < self.epsilon:
                break

    def calculate_centers(self):
        # Update the cluster centers
        um = self.membership_mat ** self.m
        self.centers = (np.dot(um.T, self.X) / np.sum(um, axis=0)[:, np.newaxis])

    def update_membership_mat(self):
        # Update the membership matrix
        dist = cdist(self.X, self.centers)
        self.old_membership_mat = self.membership_mat.copy()
        self.membership_mat = 1.0 / (dist ** (2 / (self.m - 1)))
        self.membership_mat /= np.sum(self.membership_mat, axis=1)[:, np.newaxis]

        
# Generate some random data
X = databaru

# Create an instance of the FPCM algorithm
fpcm = FPCM(n_clusters = 3, m = 2)

# Fit the algorithm to the data
fpcm.fit(X)

# Get the cluster labels for each data point
label_FPCM = np.argmax(fpcm.membership_mat, axis=1)
label_FPCM
```

```python
from sklearn.metrics import silhouette_score
import numpy as np
import skfuzzy as fuzz
from sklearn.metrics import calinski_harabasz_score
from sklearn.metrics import davies_bouldin_score

# Menghitung nilai Silhouette index
silhouette_avg_FPCM = silhouette_score(databaru, label_FPCM)
print("Silhouette index FPCM:", silhouette_avg_FPCM)
ch_score_FPCM = calinski_harabasz_score(databaru, label_FPCM)
print("CHS index FPCM:", ch_score_FPCM)
dbi_score_fpcm = davies_bouldin_score(databaru, label_FPCM)
print("dbi scroe index FPCM:", dbi_score_fpcm)
```

```python
from scipy.spatial import ConvexHull
import folium
```

```python
# Create the map object called m which is the base layer of the map
m = folium.Map(location=[df_map['Lat'].mean(), df_map['Long'].mean()],
               tiles='CartoDB positron',
               zoom_start=7)
```

```python
# create layers based on your clustering groups
layer1 = folium.FeatureGroup(name= '<u><b>Cluster 1</b></u>',show= True)
m.add_child(layer1)
layer2 = folium.FeatureGroup(name= '<u><b>Cluster 2</b></u>',show= True)
m.add_child(layer2)
layer3 = folium.FeatureGroup(name= '<u><b>Cluster 3</b></u>',show= True)
m.add_child(layer3)

#draw marker class for each group by adding CSS class
my_symbol_css_class= """ <style>
.fa-g1:before {
    font-family: Arial; 
    font-weight: bold;
    font-size: 12px;
    color: black;
    background-color:white;
    border-radius: 10px; 
    white-space: pre;
    content: ' C1 ';
    }
.fa-g2:before {
    font-family: Arial; 
    font-weight: bold;
    font-size: 12px;
    color: black;
    background-color:white;
    border-radius: 10px; 
    white-space: pre;
    content: ' C2 ';
    }
.fa-g3:before {
    font-family: Arial; 
    font-weight: bold;
    font-size: 12px;
    color: black;
    background-color:white;
    border-radius: 10px; 
    white-space: pre;
    content: ' C3 ';
    }
.fa-g1bad:before {
    font-family: Arial; 
    font-weight: bold;
    font-size: 12px;
    color: white;
    background-color:red;
    border-radius: 10px; 
    white-space: pre;
    content: ' C1 ';
    }
.fa-g2bad:before {
    font-family: Arial; 
    font-weight: bold;
    font-size: 12px;
    color: white;
    background-color:red;
    border-radius: 10px; 
    white-space: pre;
    content: ' C2 ';
    }
.fa-g3bad:before {
    font-family: Arial; 
    font-weight: bold;
    font-size: 12px;
    color: white;
    background-color:red;
    border-radius: 10px; 
    white-space: pre;
    content: ' C3 ';
    }

</style>
"""
# the below is just add above  CSS class to folium root map      
m.get_root().html.add_child(folium.Element(my_symbol_css_class))
# then we just create marker and specific your css class in icon like below
for index, row in df_map.iterrows():
    if row['cluster'] == 0:
        color='black'
        fa_symbol = 'fa-g1'
        lay = layer1
    elif row['cluster'] ==0 :
        color='black'
        fa_symbol = 'fa-g1bad'
        lay = layer1
    elif row['cluster'] == 1:
        color='purple'
        fa_symbol = 'fa-g2'
        lay = layer2
    elif row['cluster'] == 1:
        color='purple'
        fa_symbol = 'fa-g2bad'
        lay = layer2        
    elif row['cluster'] == 2:
        color='orange'
        fa_symbol = 'fa-g3'
        lay = layer3
    else:
        color='orange'
        fa_symbol = 'fa-g3bad'
        lay = layer3
        
    folium.Marker(
        location=[row['Lat'], row['Long']],
        title = row['Provinsi']+ 'group:{}'.format(str(row['Provinsi'])),
        popup = row['Provinsi']+ 'group:{}'.format(str(row['Provinsi'])),
        icon= folium.Icon(color=color, icon=fa_symbol, prefix='fa')).add_to(lay)

```

```python
#draw cluster each group
#flat line to group path
#prepare layer and color for each group
layer_list = [layer1,layer2,layer3]
color_list = ['black','purple','orange']
for g in df_map['cluster'].unique():
# this part we apply ConvexHull theory to find the boundary of each group
    # first, we have to cut the lat lon in each group 
    latlon_cut =df_map[df_map['cluster']==g].iloc[:, 1:3]
    # second, scipy already provides  the great function for ConvexHull
    # we just throw our dataframe with lat lon in this function
    hull = ConvexHull(latlon_cut.values)
    # and with magic, we can have new lat lon boundary of each group
    Lat = latlon_cut.values[hull.vertices,0]
    Long = latlon_cut.values[hull.vertices,1] 
    # the we create dataframe boundary and convert it to list of lat lon 
    # for plotting polygon in folium
    cluster = pd.DataFrame({'Lat':Lat,'Long':Long })       
    area = list(zip(cluster['Lat'],cluster['Long']))
    # plot polygon
    list_index = g-1 # minus 1 to get the same index 
    lay_cluster = layer_list[list_index ] 
    folium.Polygon(locations=area,
        color=color_list[list_index],
        weight=2,
        fill=True,
        fill_opacity=0.1,
        opacity=0.8).add_to(lay_cluster) 
            
# to let the map have selectd layer1 layer2 you created
folium.LayerControl(collapsed=False,position= 'bottomright').add_to(m)
# save it to html then we can send the file to our colleagues
m.save('mapping_cluster.html')
```
